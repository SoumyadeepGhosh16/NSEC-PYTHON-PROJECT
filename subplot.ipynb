```python
import numpy as np
import random
import matplotlib.pyplot as plt
class subplot:
    def graphplots():
        x_axis=np.random.randint(10,50,10)
        y_axis=np.random.randint(10,50,10)
        plt.subplot(2,2,1)
        plt.plot(x_axis)
        plt.plot(y_axis)

        a_axis=np.random.randint(10,70,10)
        b_axis=np.random.randint(10,70,10)
        plt.subplot(2,2,2)
        plt.barh(a_axis,b_axis)
        plt.grid()
        

        c_axis=np.random.randint(10,40,10)
        d_axis=np.random.randint(10,40,10)
        plt.subplot(2,2,3)
        plt.scatter(c_axis,d_axis)
        plt.grid()
        plt.subplot(2,2,3)

        e_axis=np.random.randint(10,70,6)
        label_axis=('A','B','C','D','E','F')
        plt.subplot(2,2,4)
        plt.pie(e_axis,labels=label_axis)
        plt.legend()
        plt.show()

subplot.graphplots()



```
